```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar datos desde el archivo CSV
df = pd.read_csv('day_wise.csv')

# Convertir la columna 'Date' a tipo datetime
df['Date'] = pd.to_datetime(df['Date'])

# Establecer 'Date' como índice
df.set_index('Date', inplace=True)

df
```

```python
# Grafico de la cantidad total de casos confirmados a lo largo del tiempo
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Confirmed'], marker='o', linestyle='-')
plt.title('Casos Confirmados de COVID-19')
plt.xlabel('Fecha')
plt.ylabel('Casos Confirmados millones')
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Grafico de nuevas muertes diarias
plt.figure(figsize=(10, 6))
plt.bar(df.index, df['New deaths'], color='red')
plt.title('Nuevas Muertes Diarias por COVID-19')
plt.xlabel('Fecha')
plt.ylabel('Nuevas Muertes')
plt.grid(axis='y')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

```

```python
# Grafico de la relación entre muertes y recuperados
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['Deaths'], label='Muertes', marker='o', linestyle='-')
plt.plot(df.index, df['Recovered'], label='Recuperados', marker='o', linestyle='-')
plt.title('Comparación de Muertes y Recuperados de COVID-19')
plt.xlabel('Fecha')
plt.ylabel('Cantidad')
plt.grid(True)
plt.legend()
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
